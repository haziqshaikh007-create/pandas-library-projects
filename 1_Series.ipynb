```python
import pandas as pd

print(pd.__version__)
```

Series in Pandas

```python
series1 = pd.Series([100,102,104])
series2 = pd.Series([100.1,102.2,104.3])
series3 = pd.Series(["a","b","c"])
series4 = pd.Series([True,False])

print(f'Int data is: \n {series1}')
print(f'Float data is: \n {series2}')
print(f'Alphabetic data: \n is {series3}')
print(f'Boolean data is: \n {series4}')
```

```python
data = [100,102,104]

series = pd .Series( data, index = ["a","b","c"])

series.loc["c"] = 200

print(series)
print(series.loc["b"])
print(series.iloc[2])
```

Filter by Values

```python
#LIST

data = [100,102,104,200,202]

series= pd.Series(data, index=["a","b","c","d","e"])

print(series[series <= 200])
```

```python
#DICTIONARY

calories = {"Day 1" : 1750,
            "Day 2" : 2100,
            "Day 3" : 1700}

series = pd.Series(calories)

series.loc["Day 3"] += 500

print(series)

print(series[series <= 2000])
```
