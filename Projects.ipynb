Student Data Analyzer

```python
import pandas as pd
```

```python
data = {
"Name" : ["Ali","Ahmed","Sara","Zain","Fatima"],
"Math" : [78,65,88,50,90],
"English" : [85,70,92,55,91],
"Science" : [90,60,95,58,89]
}

df= pd.DataFrame(data,index=[101,102,103,104,105])

print(df)
```

```python
std_avg =  df[["Math","English","Science"]].mean(axis=1)
# print("Average Marks of Each Student: ", std_avg.idxmax())
# print("Highest scorer in Maths:", df.iloc[df["Math"].argmax()])
# print("Student with Highest Overall Average:\n",df.iloc[std_avg.argmax()])
# print("\nStudent who failed:\n", df[std_avg<60])
# print("Subject wise averages: \n", df.mean(numeric_only=True))

# # New Column
# df["Total Marks"] = df[["Math","English","Science"]].sum(numeric_only=True,axis=1)
# print(df)

grade = []
for avg in std_avg:
         if avg>=90:
                 
                 grade.append("A")
         elif avg>=80:
                 
                 grade.append("B")
         elif avg>=60:
                 grade.append("C")
         else:
                 grade.append("Fail")
df["Grade"] = grade
print(df)

# print(df.info())
# print(df.describe())

print(df.sort_values("Total Marks", ascending=False).head(3))
```
